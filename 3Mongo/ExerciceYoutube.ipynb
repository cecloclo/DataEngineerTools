## Question 5

```python
cur = collection.aggregate([{"$group" : {"_id" : "$category_id", "nombre moyen de vues dans categorie" : {"$avg" : "$views"}}}])
list(cur)
```

## Question 6 

```python
cur = collection.aggregate([{"$group" : {"_id" : "$channel_title", "moyenne de likes" : {"$avg" : "$likes"}}}])
list(cur)
```
